```python
import json
import os
```

```python
def load_json(json_path):
    with open(json_path, 'r') as file:
        data = json.load(file)
    return data
```

```python
before_path = 'test_dataset/hand_landmarks.json'
before_dataset = load_json(before_path)
```

```python
before_dataset
```

```python
data = []
for i in range(len(before_dataset)):
    tmp = {}
    label = before_dataset[i]['label']
    index = before_dataset[i]['index']
    reference_point = before_dataset[i]['hand_data'][0]
    hand_data = []
    for point in before_dataset[i]['hand_data']:
        relative_x = point['x'] - reference_point['x']
        relative_y = point['y'] - reference_point['y']
        relative_z = point['z'] - reference_point['z']
        hand_data.append({'x': relative_x, 'y':relative_y, 'z':relative_z})

    tmp['label'] = label
    tmp['index'] = index
    tmp['hand_data'] = hand_data
    data.append(tmp)
print(data)
```

```python
save_path = 'before/hand_landmarks.json'
with open(save_path, 'w') as f:
    json.dump(data, f)
```
